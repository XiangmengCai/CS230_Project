```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
lr = [0.002, 0.001, 0.0005, 0.0001]
weight_decay = [0.01, 0.005, 0.0025, 0.001]

accuracy = np.array([[86.5, 88.6, 90.0, 90.9],
                     [86.9, 88.5, 92.0, 90.1],
                     [87.5, 91.1, 91.4, 91.7],
                     [80.7, 87.7, 87.9, 85.2]])

fig, ax = plt.subplots()
im = ax.imshow(accuracy, cmap='YlOrRd')

# We want to show all ticks...
ax.set_xticks(np.arange(len(weight_decay)))
ax.set_yticks(np.arange(len(lr)))
# ... and label them with the respective list entries
ax.set_xticklabels(weight_decay)
ax.set_yticklabels(lr)
ax.set_xlabel('Weight decay')
ax.set_ylabel('Learning rate')

# Create colorbar
cmap = plt.get_cmap('YlOrRd')
norm = colors.LogNorm(vmin=79.0, vmax=95.0)
im = ax.pcolormesh(weight_decay, lr, accuracy, cmap=cmap, norm=norm)
cbar = fig.colorbar(im, ax=ax, ticks=np.arange(80, 95, 2), format='%.0f')
cbar.ax.set_ylabel('Test particle accuracy', rotation=-90, va="bottom", fontsize=11)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
plt.setp(ax.get_yticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(lr)):
    for j in range(len(weight_decay)):
        text = ax.text(j, i, accuracy[i, j],
                       ha="center", va="center", color="w", fontsize=12)

# ax.set_title("Hyperparameter grid search", fontsize=14)
fig.tight_layout()
plt.show()
```
